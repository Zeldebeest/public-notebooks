This is an exploratory analysis of the various parameters used to select a subset of images from a grid-search of spotfinding parameters.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import prettyplotlib as ppl
from matplotlib import patheffects
```

I've already made this parsable by removing the top line, and all the lines between in individual images using Vim. Also changed the various numbers of white-space separators into a straight csv format.

```python
names = ['cryst_id', 'shot_no', 'resolution', 'n_refl', 'CC_ref_partial', 'CC_ref_full', 'n_refl_useda', 
         'CC_iso_partial', 'CC_iso_full', 'n_refl_used', 'a', 'b', 'c', 'alpha', 'beta', 'gamma', 'pickle_filename']
```

```python
df = pd.read_csv('/Users/zeldin/Dropbox/2014_Brunger_May17/SFC_XFEL/log_sanitized.txt', sep=',', names=names)
df = df.convert_objects(convert_numeric=True)  # make things numeric
df["pickle_filename"] = df["pickle_filename"].map(str.strip)  # remove whitespace from filenames
```

Now we need to make extra columns for height and area

```python
heights = []
areas = []
for path in df['pickle_filename']:
    heights.append(int(path[5:7]))
    areas.append(int(path[9:11]))

df['spot_area'] = areas
df['spot_height'] = heights
```

Let's also make a column for unit cell distance.

```python
from cctbx.uctbx.determine_unit_cell import NCDist
from xfel.clustering.singleframe import SingleFrame as SF

g6_target = SF.make_g6([68.56899236567753, 169.230375072626, 288.16104199650215, 90.0, 90.0, 90.0 ])

def g6_and_dist(a, b, c, alpha, beta, gamma):
    uc = (a, b, c, alpha, beta, gamma)
    g6 = SF.make_g6(uc)
    return NCDist(g6, g6_target)

df['dist_from_long'] = np.vectorize(g6_and_dist)(df['a'], df['b'], df['c'], df['alpha'], df['beta'], df['gamma'])
df['dist_from_long'].hist(bins=50)
len(df[df['dist_from_long'] > 10000])
```

**ahhh!!** there are clearly some outliers still here... Let's filter at 10k

```python
df = df[df['dist_from_long'] < 10000]
df['dist_from_long'].hist()
```

```python
df.hist(column=['spot_area', 'spot_height'], bins=np.arange(0.5, 10.5, 1), sharex=True, sharey=True, 
        figsize=(20,6), xlabelsize=20)
_ = plt.xticks(np.arange(1, 10, 1.0))
```

Create unique key for each image

## Create a list of things in which we are intersted

```python
vals = ['CC_ref_full', 'CC_ref_partial', 'CC_iso_full', 'CC_iso_partial', 'resolution', 'n_refl_used', 'dist_from_long']
```

```python
gdf = df.groupby(['spot_height', 'spot_area'], as_index=False).mean()  # Group by spot params
```

```python
for val in vals:
    mtrz = np.zeros((10,10))
    val_mean = gdf[val].mean()
    for height, area, v in gdf[['spot_height', 'spot_area', val]].as_matrix():
        mtrz[height-1, area-1] = v - val_mean
    fig, ax = plt.subplots(1, figsize=(12,8))
    ppl.pcolormesh(fig, ax, mtrz, 
                   xticklabels=range(1, 11), 
                   yticklabels=range(1,11))
    for i in range(10):
        for j in range(10):
            ax.text(j + 0.5, i + 0.5, '{:.1}'.format(mtrz[i, j]),
                    size='medium', ha='center', va='center',
                    path_effects=[patheffects.withSimplePatchShadow(shadow_rgbFace=(1,1,1))])
    plt.xlabel("Spot Area")
    plt.ylabel("Spot Height")
    plt.title("{}: Color value represent deviation from mean ({:.2})".format(val, val_mean), size='large')
```

## Scatter matrix to look for correlations

```python
h_a = ['spot_height', 'spot_area']
h_a.extend(vals)
h_a.remove('CC_iso_partial')  # partial and full are highly correlated.
h_a.remove('CC_ref_partial')   # partial and full are highly correlated.
_ = pd.scatter_matrix(gdf[h_a], figsize=(22, 22), hist_kwds={'bins':20})
```

```python
gdf[h_a].corr()
```

## Are height and area independent?
We fit the model $val = \theta_0 + \theta_1 area + \theta_2 height + \theta_3 height\times area$.

Overall, height and area seem to have a fairly independent effect on CC's. 

```python
for val in vals:
    print val 
    print pd.ols(y=gdf[val], x={'height': gdf['spot_height'], 'area': gdf['spot_area'], 
                                'interaction': gdf['spot_height'] * gdf['spot_area']})
    print "\n\n\n\n"
```
